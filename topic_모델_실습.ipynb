```
import numpy as np
```

```
doc_ls = [
          '바나나 사과 포도 포도 짜장면',
          '사과 포도',
          '포도 바나나',
          '짜장면 짬뽕 탕수육',
          '볶음밥 탕수육',
          '짜장면 짬뽕',
          '라면 스시',
          '스시 짜장면',
          '가츠동 스시 소바',
          '된장찌개 김치찌개 김치',
          '김치 된장 짜장면',
          '비빔밥 김치'
          ]

doc_list=[]
for i in range(len(doc_ls)):
    doc_list.append(doc_ls[i].split())
doc_list

# 토큰 생성 
tokens = []
tmp = []
for i in range(len(doc_list)):
    tmp += doc_list[i]
tokens = list(set(tmp))
tokens

# 토큰 인덱싱
mapping_w2i = {}
mapping_i2w = {}
for i, v in enumerate(tokens):
    mapping_w2i[v] = i
    mapping_i2w[i] = v
tokens_w2i = mapping_w2i
tokens_i2w = mapping_i2w
print(tokens_w2i)
print(tokens_i2w)



# TDM  ( text가 행 단어횟수가 열 -> 텍스트당 단어 등장수 ) = 문서당 토큰 카운트
doc_cnt = np.zeros((len(doc_ls), len(tokens)))
doc_cnt


for i in range(len(doc_ls)):
    for j in range(len(tokens)):
        if tokens[j] in doc_list[i]: # ※주의 : in을 쓰면 한 문서에 2번이상 나오는 것도 1로 찍힘
            doc_cnt[i][j] += doc_list[i].count(tokens[j])
doc_cnt

```

```
docs = [d.split() for d in doc_ls]

from collections import defaultdict
word2id = defaultdict(lambda : len(word2id))
[word2id[w] for d in docs for w in d ]

```
